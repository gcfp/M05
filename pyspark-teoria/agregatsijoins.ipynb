```python
# Importem les llibreries necessàries
from pyspark.sql import SparkSession

# Creem la sessió de Spark
spark = SparkSession.builder.getOrCreate()

# Carreguem les dades
vols = spark.read.csv('flights_small.csv', header=True)
avions = spark.read.csv('planes.csv', header=True)
aerolines = spark.read.csv('airlines.csv', header=True)

# Creem les vistes temporals
vols.createOrReplaceTempView('vols')
avions.createOrReplaceTempView('avions')
aerolines.createOrReplaceTempView('aerolines')

# 1. Quin és l'avió («tailnum») amb major petjada de carboni.
query = "SELECT tailnum, SUM(air_time) AS petjada_carboni FROM vols GROUP BY tailnum ORDER BY petjada_carboni DESC LIMIT 1"
spark.sql(query).show()

# 2. Quin és l'avió («tailnum») amb menor petjada de carboni. ¿ha estat temps sense volar?
query = "SELECT tailnum, SUM(air_time) AS petjada_carboni FROM vols GROUP BY tailnum ORDER BY petjada_carboni ASC LIMIT 1"
spark.sql(query).show()

# 3. Quina és la petjada promig i la petjada mitjana segons «tailnum». Si hi aprecieu diferència notable a què ho atribuiu?
query = "SELECT tailnum, AVG(air_time) AS petjada_carboni_promig, AVG(air_time) AS petjada_carboni_mitjana FROM vols GROUP BY tailnum"
spark.sql(query).show()

# 4. Quina és l'aerolinia amb major petjada de carboni.
query = "SELECT airline, SUM(air_time) AS petjada_carboni FROM vols GROUP BY airline ORDER BY petjada_carboni DESC LIMIT 1"
spark.sql(query).show()

# 5. Quina és l'aerolinia amb menor petjada de carboni.
query = "SELECT airline, SUM(air_time) AS petjada_carboni FROM vols GROUP BY airline ORDER BY petjada_carboni ASC LIMIT 1"
spark.sql(query).show()

# 6. Quin és el fabricant d'avions amb major petjada de carboni.
query = "SELECT manufacturer, SUM(air_time) AS petjada_carboni FROM vols JOIN avions ON vols.tailnum = avions.tailnum GROUP BY manufacturer ORDER BY petjada_carboni DESC LIMIT 1"
spark.sql(query).show()

# 7. Quin és el fabricants d'avions amb menor petjada de carboni.
query = "SELECT manufacturer, SUM(air_time) AS petjada_carboni FROM vols JOIN avions ON vols.tailnum = avions.tailnum GROUP BY manufacturer ORDER BY petjada_carboni ASC LIMIT 1"
spark.sql(query).show()

# 8. Quin és el model d'avions amb major petjada de carboni.
query = "SELECT model, SUM(air_time) AS petjada_carboni FROM vols JOIN avions ON vols.tailnum = avions.tailnum GROUP BY model ORDER BY petjada_carboni DESC LIMIT 1"
spark.sql(query).show()

# 9. Quin és el model d'avions amb menor petjada de carboni.
query = "SELECT model, SUM(air_time) AS petjada_carboni FROM vols JOIN avions ON vols.tailnum = avions.tailnum GROUP BY model ORDER BY petjada_carboni ASC LIMIT 1"
spark.sql(query).show()

# 10. Fes una taula de petjada de carboni en funció dels anys de fabricació dels avions
query = "SELECT year, SUM(air_time) AS petjada_carboni FROM vols JOIN avions ON vols.tailnum = avions.tailnum GROUP BY year ORDER BY year"
spark.sql(query).show()

# 11. Quina és l'aerolinia amb major petjada de carboni normalitzat al número de seients disponible a cada vol.
query = "SELECT airline, SUM(air_time) / COUNT(seats) AS petjada_carboni FROM vols JOIN avions ON vols.tailnum = avions.tailnum GROUP BY airline ORDER BY petjada_carboni DESC LIMIT 1"
spark.sql(query).show()

# 12. Quina és l'aerolinia amb menor petjada de carboni normalitzat al número de seients disponible a cada vol.
query = "SELECT airline, SUM(air_time) / COUNT(seats) AS petjada_carboni FROM vols JOIN avions ON vols.tailnum = avions.tailnum GROUP BY airline ORDER BY petjada_carboni ASC LIMIT 1"
spark.sql(query).show()

# 13. Quin és el fabricant d'avions amb major petjada de carboni normalitzat al número de seients disponible a cada vol.
query = "SELECT manufacturer, SUM(air_time) / COUNT(seats) AS petjada_carboni FROM vols JOIN avions ON vols.tailnum = avions.tailnum GROUP BY manufacturer ORDER BY petjada_carboni DESC LIMIT 1"
spark.sql(query).show()

# 14. Quin és el fabricants d'avions amb menor petjada de carboni normalitzat al número de seients disponible a cada vol.
query = "SELECT manufacturer, SUM(air_time) / COUNT(seats) AS petjada_carboni FROM vols JOIN avions ON vols.tailnum = avions.tailnum GROUP BY manufacturer ORDER BY petjada_carboni ASC LIMIT 1"
spark.sql(query).show()

# 15. Quin és el model d'avions amb major petjada de carboni normalitzat al número de seients disponible a cada vol.
query = "SELECT model, SUM(air_time) / COUNT(seats) AS petjada_carboni FROM vols JOIN avions ON vols.tailnum = avions.tailnum GROUP BY model ORDER BY petjada_carboni DESC LIMIT 1"
spark.sql(query).show()

# 16. Quin és el model d'avions amb menor petjada de carboni normalitzat al número de seients disponible a cada vol.
query = "SELECT model, SUM(air_time) / COUNT(seats) AS petjada_carboni FROM vols JOIN avions ON vols.tailnum = avions.tailnum GROUP BY model ORDER BY petjada_carboni ASC LIMIT 1"
spark.sql(query).show()

# 17. Fes una taula de petjada de carboni en funció dels anys de fabricació dels avions normalitzat al número de seients disponible a cada vol.
query = "SELECT year, SUM(air_time) / COUNT(seats) AS petjada_carboni FROM vols JOIN avions ON vols.tailnum = avions.tailnum GROUP BY year ORDER BY year"
spark.sql(query).show()

# 18. Quina és l'aerolinia amb major petjada de carboni normalitzat al número de passatgers transportats a cada vol.
query = "SELECT airline, SUM(air_time) / COUNT(passengers) AS petjada_carboni FROM vols JOIN avions ON vols.tailnum = avions.tailnum GROUP BY airline ORDER BY petjada_carboni DESC LIMIT 1"
spark.sql(query).show()

# 19. Quina és l'aerolinia amb menor petjada de carboni normalitzat al número de passatgers transportats a cada vol.
query = "SELECT airline, SUM(air_time) / COUNT(passengers) AS petjada_carboni FROM vols JOIN avions ON vols.tailnum = avions.tailnum GROUP BY airline ORDER BY petjada_carboni ASC LIMIT 1"
spark.sql(query).show()
```
